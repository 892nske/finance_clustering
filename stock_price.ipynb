```python
import pandas as pd
import yfinance as yf
```

```python
symbol_list = pd.read_csv('company_nyse.csv',index_col=0)
```

```python
symbol_list = symbol_list[0:100]
```

```python
start_date = '2023-01-01'
end_date = '2023-12-31'
```

```python
symbol = symbol_list['Symbol'][0]
stock = yf.Ticker(symbol)
stock_data = stock.history(start=start_date, end=end_date)
```

```python
r = stock_data['Close'].pct_change()
r = r.dropna()
```

```python
r_mean = r.mean()
r_std = r.std()
```

```python
symbol_list['Return'] = 0
symbol_list['STD'] = 0

for row in symbol_list.itertuples():
    symbol = row.Symbol
    stock = yf.Ticker(symbol)
    stock_data = stock.history(start=start_date, end=end_date)
    r = stock_data['Close'].pct_change()
    r = r.dropna()
    r_mean = r.mean()
    r_std = r.std()
    print(symbol + ':')
    print(r_mean)
```
